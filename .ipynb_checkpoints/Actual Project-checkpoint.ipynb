EXPLORATORY ANALYSIS. 

The Problem: Can we predict UNS with the variables PEG and LPR, using KNN classification. 

*** Rough Documentation of Code:

- importing data set
- vizualization of the problem (we see that classification is possible, because there seems to be clusters of data surrounding the different levels of knowledge along with UNS

```r
#load necessary packages
library(tidyverse) #load tidyverse - needed for plots
library(tidymodels) #load tidymodels - needed for KNN classification
library(readxl) #load readxl - needed to load excel file
library(dplyr) #load dplyr - needed for data frame manipulation
```

```r
#load the data by Kahraman et al. from the following URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00257/Data_User_Modeling_Dataset_Hamdi%20Tolga%20KAHRAMAN.xls"

#download
download.file(url, "data.xls")

#sort by the training and testing data
training_data <- read_excel("data.xls", sheet = 2)
testing_data <- read_excel("data.xls", sheet = 3)
```

```r
#create a training data set and scale by a factor of 10
training_select <- training_data %>% 
                select(LPR, PEG, UNS) %>%
                mutate(LPR_scaled = LPR * 10, PEG_scaled = PEG * 10)

#### WHY are we scaling the data by a factor of 10???

#how many observations?
number_obvs <- nrow(training_select)
print("Number of observations in each class:")
number_obvs


#how much missing data? (how many NAs are there?)
missing_data <- data.frame (sum(is.na(training_select$PEG)),
    sum(is.na(training_select$LPR)),
    sum(is.na(training_select$UNS)))

colnames(missing_data) <- c('missing_PEG','missing_LPR','missing_UNS')
missing_data

#add 2 columns: the average of performance on both exam types, and the scaled average
ave_exam_performance = rowMeans(training_select[1:2]) #taking the average
training_select <- training_select %>%
    mutate(ave_exam_performance = ave_exam_performance, ave_exam_performance_scaled = ave_exam_performance*10) 


#create reproducable randomization
set.seed(123)

#choose 20 rows to view in a table - gives an idea of what the training data looks like
#shows that data is already tidy
index = sample(1:number_obvs, 20, replace = TRUE)
training_select[index,]
```

```r
#plot the training data: PEG vs LPR, colour coded by knowledge level
options(repr.plot.width = 10, repr.plot.height = 8)
training_plot <- training_select %>%
        ggplot(aes(x = LPR_scaled, y = PEG_scaled)) + 
        geom_point(aes(colour = UNS)) + 
        labs(x = "Related Object Exam Performance", y = "Goal Object Exam Performance", colour = "Knowledge Level") +                             
        ggtitle("Initial Visualization of Data (scaled)")+
        theme(text = element_text(size = 15))
training_plot
```

```r
#plot the training data: average exam peformance, colour coded by knowledge level
options(repr.plot.width = 10, repr.plot.height = 8)
training_plot2 <- training_select %>%
        ggplot(aes(x = ave_exam_performance_scaled, fill = UNS)) + 
        geom_histogram()+
        labs(x = "Average Exam Performance") +                                                          
        ggtitle("Average Exam Performance (scaled) and Knowledge Level")+
        theme(text = element_text(size = 15))

training_plot2
```

```r
#wrangling the testing data, to fit our purposes (no preprocessing done here)

UM_testing <- testing_data %>%
                select(LPR:UNS) %>%
                mutate(UNS = as_factor(UNS))
#UM_testing

# getting rid of unnecessary columns

UM_training <- training_select %>%
                select(LPR: UNS)

# making the outcome variable into a factor variable

UM_training <- UM_training %>%
                mutate(UNS = as_factor(UNS))

#UM_training

# checking the different categories in UNS
# output (very_low, low, high, middle)

levels_training <- UM_training %>%
                    pull(UNS) %>%
                    levels()

#levels_training


```

```r
### BUILDING THE WORKFLOW

# we are testing K values 1 - 20

k_vals <- tibble(neighbors = seq(from = 1, to = 20, by = 1))
k_vals

# we are performing 10 fold cross-validation

UM_vfold <- vfold_cv(UM_training, v = 10, strata = UNS)

#pre-processing the data, by standardizing the predictors 

UM_recipe <- recipe(UNS~., data = UM_training) %>%
             step_scale(all_predictors()) %>%
             step_center(all_predictors())

###NOTE TO OTHERS: SHOULD we include prep(), bake() to show that we checked whether the recipe was working?

#creating the model specification, tuning for K

UM_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) %>%
            set_engine("kknn") %>%
            set_mode("classification")


# putting it all together in the workflow and getting the results

UM_workflow <- workflow() %>%
                add_recipe(UM_recipe) %>%
                add_model(UM_spec) %>%
                tune_grid(resamples = UM_vfold, grid = k_vals) %>%
                collect_metrics()

UM_workflow
```

METRICS COLLECTED

This bit is to collect the data and find the best_k and assign a min value

```r
# Filtering out just the accuracies

UM_accuracies <- UM_workflow %>%
                  filter(.metric == "accuracy") %>%
                  select(neighbors, mean)
#UM_accuracies

#putting it into a vizualization


UM_accuracy_plot <- UM_accuracies %>%
                    ggplot(aes(x = neighbors, y = mean)) +
                    geom_point() +
                    geom_line() +
                    scale_x_continuous(breaks = 1:20) +
                    labs(x = "K value", y = "Accuracy Estimate") +
                    ggtitle("Accuracy Estimate for each K value (1 - 20)") +
                    theme(text = element_text(size = 20))
UM_accuracy_plot

### IMPORTANT NOTE: the accuracies for 7 and 8 are identical to 6 decimal places. I've chosen the lower one arbitrarily, we should probably check this again

# finding the maximum K value (Max k = 7). k_max is a tibble with the best k value and % accuracy. % accuracy = 95.67% accuracy for K = 7
k_max <- UM_accuracies %>%
            filter(mean == max(mean)) %>%
            slice(1) %>%
            mutate(mean = mean * 100)
k_max

#ive put the k_max_num into a numerical variable, so we can easily put into the code evaluating our testing data

k_max_num <- UM_accuracies %>%
             filter(mean == max(mean)) %>%
             slice(1) %>%
             select(neighbors) %>%
             pull()
k_max_num

```
